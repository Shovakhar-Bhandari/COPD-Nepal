# Extract the relevant columns

```python
csv_selected_columns = csv_data[['sex', 'bmi', 'age', 'location', 'smoke']]
excel_selected_columns = excel_file[['Heigt(cm)', 'weight kg', 'BMI']]

excel_selected_columns.rename(columns={"Heigt(cm)": "Height(cm)", "weight kg": "weight_kg"}, inplace=True)


# Merged data
merged_data = pd.merge(csv_selected_columns, excel_selected_columns, 
                       left_on="bmi", right_on="BMI", how="inner")
```
